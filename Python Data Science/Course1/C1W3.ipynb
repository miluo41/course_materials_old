```python
import numpy as np
```

```python
def min_max(row):
    data=row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min':np.min(data),'max':np.max(data)})
```

```python
df.apply(min_max,axis=1).head()
```

```python
#below is the case to add new column to the original df
def min_max2(row):
    data=row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max']=np.max(data)
    row['min']=np.min(data)
    return row
    
df.apply(min_max2,axis=1)
```

```python

rows = ['POPESTIMATE2010',
        'POPESTIMATE2011',
        'POPESTIMATE2012',
        'POPESTIMATE2013',
        'POPESTIMATE2014',
        'POPESTIMATE2015']

df.apply(lambda x:np.max(x[rows]),axis=1)
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('census.csv')
```

```python
df=df[df['SUMLEV']==50]
```

```python
df.head()
```

```python
%%timeit
for state in df['STNAME'].unique():
    ave=np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('average of '+state+str(ave))
```

```python
%%timeit
for group, frame in df.groupby('STNAME'):
    ave=np.average(frame['CENSUS2010POP'])
    print('average of '+group+str(ave))
```

```python
#using a function on group by, the function is operated on the row index
df=pd.read_csv('census.csv')
df=df.set_index('STNAME')
def fun(x):
    if x[0]<'M':
        return 0
    if x[0]<'Q':
        return 1
    return 2
for group, frame in df.groupby(fun):
    print(str(group)+'has'+str(len(frame)))
```
