```python
import matplotlib.gridspec as gridspec
```

```python
x=np.random.random(size=10000)
y=np.random.normal(loc=0,scale=1,size=10000)
```

```python
plt.figure()
gspec=gridspec.GridSpec(3,3)
top=plt.subplot(gspec[0,1:])
left=plt.subplot(gspec[1:,0])
main=plt.subplot(gspec[1:,1:])
top.hist(x,bins=1000,normed=True)
left.hist(y,bins=1000,normed=True,orientation='horizontal')
left.invert_xaxis()
main.scatter(x,y,s=0.1)
```

```python
import matplotlib.animation as animation
```

```python
n=100
x=np.random.randn(n)
```

```python
def update(curr):
    if curr==n:
        a.event_scource.stop()
    plt.cla()
    bins=np.arange(-4,4,0.5)
    plt.hist(x[:curr],bins=bins)
    plt.axis=([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_ylabel('Value')
    plt.annotate('n={}'.format(curr),[3,27])
```

```python
fig=plt.figure()
a=animation.FuncAnimation(fig, update, interval = 100)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
plt.figure()
data=np.random.random(size=10)
plt.plot(data)
def on_click(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixel {}, {} and data {}, {}'.format(event.x,event.y,'\n',
                                                                       event.xdata,event.ydata))
plt.gcf().canvas.mpl_connect('button_press_event',on_click)
```

```python
origins=['China','Brazil','India','USA','Canada','UK','Germany','Iraq',
         'Chile','Mexico']
```

```python
np.random.shuffle(origins)
```

```python
plt.figure(figsize=(5,5))
height=np.random.rand(10)
weight=np.random.rand(10)
df=pd.DataFrame({'origins':origins,'height':height,'weight':weight})
plt.scatter(df.height,df.weight,picker=5)
plt.gca().set_xlabel('Weight')
plt.gca().set_ylabel('Height')
```

```python
def onpick(event):
    origin=df.iloc[event.ind[0]]['origins']
    plt.gca().set_title('selected sample come from {}'.format(origin))
plt.gcf().canvas.mpl_connect('pick_event',onpick)
```
