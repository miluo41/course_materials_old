#### minimization using simplex method

```python
## the .dot method in np is interesting. If 2D array .dot a 1-D vecor, 
## the 1D vector is interpreted to be a column vector by default.
## np is good at propagation, thus there is usually no need to explicitly reshape the 1D array into a column vector.
## in the function below, y, and theta are 1-D vectors.
```

```python
def compute_cost(theta,X,y,mylambda=0):
    """
    theta is 1-D np.array in n dimension,
    reshape is needed for vectorized implementation
    """
    n=X.shape[1]
    hx=sigmoid(X.dot(theta))
    term_1=-y.dot(np.log(hx))
    term_2=-(1-y).dot(np.log(1-hx))
    reg_term=theta.dot(theta)*mylambda/2
    J=float((1/m)*(term_1+term_2+reg_term))
    return J
```

```python
theta_ini=np.array([0,0,0])
compute_cost(theta_ini,X,y)
```

```python
compute_cost(theta_ini,X,y,10000)
```

```python
## In the scipy.optimize.minimize method options, the common choices of "CG" and "BFSG" require the gradient function (jac),
## but 'Nelder-Mead' method, which uses simplex requires no gradient term. 
## https://docs.scipy.org/doc/scipy-0.19.0/reference/optimize.minimize-neldermead.html#optimize-minimize-neldermead
```

```python
res=minimize(compute_cost,theta_ini,args=(X,y),method='Nelder-Mead',options={'maxiter':400})
```

#### minimization using BFGS and CG method

```python
def compute_grad(theta,X,y,mylambda=0):
    """
    theta is 1-D np.array in n dimension,
    reshape is needed for vectorized implementation
    """
    m=X.shape[0]
    n=X.shape[1]
    theta_reg=theta.copy()
    theta_reg[0]=0
    hx=sigmoid(X.dot(theta))
    grad=(1/m)*X.T.dot((hx-y))+(mylambda/m)*theta_reg
    return grad
```

```python
res2=minimize(compute_cost,theta_ini,args=(X,y),method='CG',jac=compute_grad)
```

```python
res3=minimize(compute_cost,theta_ini,args=(X,y),method='BFGS',jac=compute_grad)
```

```python
def feature_mapping(x1,x2,degree):
    """
    x1 and x2 are mx1 vectors
    output is the polynomial feature matrix of x1 and x2
    from zero degree, to a total power determined by 'degree' arg
    """
    assert len(x1)==len(x2)
    m=len(x1)
    tmp=[i for i in range(degree+2)]
    n=sum(tmp)
    feature_matrix=np.zeros((m,n))
    idx=0
    for i in range(degree+1):
        for j in range(degree-i+1):
            feature_col=(x1**i)*(x2**(j))
            feature_matrix[:,idx]=feature_col
            idx+=1
    return feature_matrix
```

```python
x1=X[:,1]
x2=X[:,2]
f_mat=feature_mapping(x1,x2,6)
```

```python
def optimize_theta(theta_0,X,y,mylambda=0):
    results=minimize(compute_cost,theta_0,args=(X,y,mylambda),method='BFGS',jac=compute_grad)
    return results.x,results.fun
```
