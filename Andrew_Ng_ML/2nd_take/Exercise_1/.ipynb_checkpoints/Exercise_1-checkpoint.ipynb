#### 2.2 Gradient descent

```python
theta=np.zeros((2,1))
iterations=1500
alpha=0.01
```

```python
def compute_cost(theta,X,y):
    """
    theta is the coefficient matrix of dimension n x 1,
    where n is the number of features.
    compute the cost function J
    """
    diff=X.dot(theta)-y
    m=X.shape[0]
    J=float(1.0/(2*m)*diff.T.dot(diff))
    return J
```

```python
compute_cost(theta,X,y)
```

```python
def compute_grad(theta,X,y):
    """
    theta is the input coefficient matrix of n x 1,
    compute and return the gradient of theta of the same dimension.
    """
    m=X.shape[0]
    diff=X.dot(theta)-y
    grad=1.0/m*(X.T.dot(diff))
    return grad
```

```python
compute_grad(theta,X,y)
```

```python
def compute_grad_desc(theta,X,y,alpha=0.01,iterations=1500):
    jvec=[]
    theta_hist=[]
    for i in range(iterations):
        grad=compute_grad(theta,X,y)
        theta=theta-alpha*grad
        jvec.append(compute_cost(theta,X,y))
        theta_hist.append(list(theta[:,0]))
    return theta,jvec,theta_hist
```

```python
theta_ini=np.zeros((X.shape[1],1))
```

```python
theta,jvec,theta_hist=compute_grad_desc(theta_ini,X,y)
```

```python
def plot_convergence(jvec):
    plt.figure(figsize=(10,6))
    plt.plot(range(len(jvec)),jvec,'bo',ms=3,alpha=0.5)
    plt.xlabel('number of iterations',fontsize=15)
    plt.ylabel(r'J($\theta$)',fontsize=15)
    plt.title('Convergence Plot',fontsize=15)
    plt.grid(True)
```

```python
plot_convergence(jvec)
```

```python
def myfit(theta,X):
    prediction=theta[0]+theta[1]*X[:,1]
    prediction=prediction.reshape((X.shape[0],1))
    return prediction
```

```python
hx=myfit(theta,X)
```

```python
plt.figure(figsize=(10,6))
plt.plot(X[:,1],y[:,0],'rx',ms=10,label='Training Data')
plt.plot(X[:,1],hx,'b-',label='Hypothesis: %0.2f+%0.2f*x1'%(theta[0],theta[1]))
plt.legend()
plt.xlabel('Profit in $10,000s',fontsize=15)
plt.ylabel('Population of City in 10,000s',fontsize=15)
plt.grid(True)
```

#### 3.1 Feature nomalization 

```python
my_data2=np.loadtxt('ex1data2.txt',delimiter=',')
```

```python
my_data2.shape
```

```python
X2=my_data2[:,:-1]
y2=my_data2[:,[-1]]
X2.shape,y2.shape
```

```python
X2=np.insert(X2,0,1,axis=1)
X2.shape
```

```python
def normalize_feature(X):
    x0=X[:,[0]]
    X_rest=X[:,1:]
    X_mean=np.apply_along_axis(np.mean,0,X_rest)
    X_sd=np.apply_along_axis(np.std,0,X_rest)
    X_n=(X_rest-X_mean)/X_sd
    return np.column_stack((x0,X_n)),X_mean,X_sd
```

```python
X2_norm,X_mean,X_sd=normalize_feature(X2)
```

```python
theta_ini=np.zeros((X2_norm.shape[1],1))
```

```python
theta,jvec,theta_hist=compute_grad_desc(theta_ini,X2_norm,y2)
```
